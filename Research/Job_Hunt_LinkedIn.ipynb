```python

#get all the html details
def scrape_linkedin_jobs(jobtitle = None, location = None):
    base_site = "https://in.linkedin.com/jobs/search?"
    #forming the complete site
    if jobtitle != None and location != None:
        site = base_site + "keywords=" + str(jobtitle) + "&location=" + str(location) + "&position=1&pageNum=0"
    elif jobtitle != None:
        site = base_site + "keywords=" + str(jobtitle) + "&position=1&pageNum=0"
    elif location!= None:
        site = base_site + "&location=" + str(location) + "&position=1&pageNum=0"
    else:
        site = base_site + "&position=1&pageNum=0"
    
    #inititating the driver
    driver = webdriver.Firefox()
    print(site)
    driver.get(site)
    #sleep to load the website properly
    time.sleep(3)
    elem = driver.find_element(By.TAG_NAME, "body")
    #print('elem',elem, '\n')
    no_of_pagedowns = 10
    
    #scroll logic and not executing 'On-Click' event as it will give large results
    #working with small details
    length = 0
    while no_of_pagedowns:
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)
        no_of_pagedowns-=1
        soup = BeautifulSoup(driver.page_source,"html.parser")
        results = soup.find(class_='jobs-search__results-list')
        
        if length < len(results):
            length = len(results)
        else:
            break
    
    
            
    driver.close()
    
    df = load_data_dataframe(results)
    
    return df
        
        
def load_data_dataframe(results = None):
    list_data = []
    if results == None:
        print("Something is wrong. Kindly check with administrator")
    else:
        for li in results.find_all('li'):

            #URL to qpply job
            URL = li.find('a').get('href')
            print(URL)
            #get job-title
            title = li.find('a').get_text()
            print(title)


            #get company-name
            company = li.find('h4').get_text()
            print(company)

            #get job location
            location = li.find('span', class_ = 'job-search-card__location').get_text()
            print(location)

            #Hiring Status
            hiring_status = li.find('span', class_ = 'job-posting-benefits__text')
            if hiring_status is None:
                hs = ' '
                pass
            else:
                hs = hiring_status.text
                print(hiring_status.text)

            #listed date
            list_date = li.find('time', class_ = 'job-search-card__listdate')
            if list_date is None:
                ld = ' '
                pass
            else:
                ld  = list_date.text
                print(list_date.text)
                
            #   Appending data to the DataFrame
            list_data.append(['LinkedIn',URL,title,company,location,hs,ld])
        
    df=pd.DataFrame(list_data, columns=['SOURCE','URL','TITLE','COMPANY','LOCATION','HIRING_STATUS','LIST_DATE'])
    return df




```
