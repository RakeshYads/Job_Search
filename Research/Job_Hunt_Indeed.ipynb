```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
site = "https://in.indeed.com/jobs?q=data+scientist&l=Hyderabad&start=00"
```

```python
driver = webdriver.Firefox()
driver.get(site)
time.sleep(3)
soup = BeautifulSoup(driver.page_source,"html.parser")
print(soup.prettify())
```

```python
results = soup.find(class_='css-zu9cdh eu4oa1w0')
print(results)
```

```python

for li in results.find_all('li', class_ = 'css-5lfssm eu4oa1w0'):
    
    #URL to qpply job
    if li.find('div', class_ = 'mosaic mosaic-empty-zone nonJobContent-desktop'):
        pass
    else:
        URL = li.find('a')
        if URL is None:
            pass
        else:
            url = 'https://in.indeed.com' + URL.get('href')
            print(url)
        #get job-title

        title = li.find('a').get_text()
        print(title)


        #get company-name
        company = li.find('span', class_ = 'css-63koeb eu4oa1w0').get_text()
        print(company)

        #get job location
        location = li.find('div', class_ = 'css-1p0sjhy eu4oa1w0').get_text()
        print(location)


        #listed date
        list_date = li.find('span', class_ = 'css-10pe3me eu4oa1w0')
        if list_date is None:
            pass
        else:
            print(list_date.text)

```
